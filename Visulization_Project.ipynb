#### 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- These libraries are used to load the dataset and create visualizations for data analysis.

#### 2. Load Dataset

```python
df = pd.read_csv("spotify_songs_visualization_dataset.csv")
```

- The dataset is loaded using the `read_csv()` function from the Pandas library to read the Spotify songs CSV file into a DataFrame.

#### 4. Visualization Operations

##### 1. Popularity Trend of Songs (Line plot)

```python
plt.figure(figsize=(8,5))

plt.plot(df["popularity"])

plt.title("Popularity Trend of Songs")

plt.xlabel("Song Index")
plt.ylabel("Popularity")

plt.show()
```

- A line plot is used to visualize the trend of song popularity across the dataset.  
- The X-axis represents the song index, while the Y-axis represents the popularity score.  
- This plot helps observe how popularity values vary among different songs.

##### 2. Genre Distribution (Bar plot)

```python
genre_count = df["genre"].value_counts()

plt.figure(figsize=(8,5))

plt.bar(genre_count.index, genre_count.values)

plt.xlabel("Genre")
plt.ylabel("Number of Songs")

plt.xticks(rotation=45)

plt.title("Genre Distribution")

plt.show()
```

- A bar plot is used to show the number of songs in each music genre.  
- The X-axis represents different genres, and the Y-axis represents the number of songs.  
- This visualization helps identify which genres are most common in the dataset.

##### 3. Popularity Distribution of Songs (Histogram)

```python
plt.figure(figsize=(8,5))

plt.hist(df["popularity"], bins=20)

plt.title("Popularity Distribution")

plt.xlabel("Popularity")
plt.ylabel("Frequency")

plt.show()
```

- A histogram is used to visualize the distribution of song popularity scores.  
- The X-axis represents the popularity score, and the Y-axis shows the number of songs (frequency).  
- This plot helps understand how popularity values are spread across the dataset.

##### 4. Danceability vs Energy Relationship (Scatter Plot)

```python
plt.figure(figsize=(8,5))

plt.scatter(df["danceability"], df["energy"])

plt.title("Danceability vs Energy")

plt.xlabel("Danceability")
plt.ylabel("Energy")

plt.show()
```

- A scatter plot is used to show the relationship between danceability and energy of songs.  
- The X-axis represents danceability and the Y-axis represents energy.  
- This visualization helps identify whether songs with higher danceability tend to have higher energy.

##### 5. Number of Songs by Genre (Count plot)

```python
plt.figure(figsize=(8,5))

sns.countplot(y="genre", data=df)

plt.title("Count of Songs by Genre")

plt.show()
```

- A count plot is used to display the number of songs in each genre category.  
- The Y-axis represents different music genres, while the X-axis shows the count of songs.  
- This visualization helps identify which genres appear most frequently in the dataset.

##### 6. Popularity by Genre (Box plot)

```python
plt.figure(figsize=(10,5))

sns.boxplot(x="genre", y="popularity", data=df)

plt.xticks(rotation=45)

plt.title("Popularity by Genre")

plt.show()
```

- A box plot is used to compare the distribution of song popularity across different genres.  
- The X-axis represents music genres and the Y-axis represents the popularity score.  
- This visualization helps identify variations, median popularity, and outliers among genres.

##### 7. Feature Correlation (Heatmap)

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")

plt.show()
```

- A heatmap is used to visualize the correlation between numerical features in the dataset.  
- It helps identify relationships between variables such as popularity, danceability, energy, tempo, and loudness.  
- Higher correlation values indicate stronger relationships between features.

##### 8. Relationship Between Numerical Features (Pair plot)

```python
sns.pairplot(df[["popularity","danceability","energy","tempo"]])

plt.show()
```


- A pair plot is used to visualize relationships between multiple numerical features in the dataset.  
- It shows scatter plots and distributions for variables like popularity, danceability, energy, and tempo.  
- This visualization helps identify patterns and correlations between different song features.
